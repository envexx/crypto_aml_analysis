```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import plotly.express as px
import plotly.graph_objects as go
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
# Configure plots
%matplotlib inline
plt.style.use('ggplot')
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Read the processed data
processed_tx = pd.read_csv('../data/processed/processed_transactions.csv')
user_metrics = pd.read_csv('../data/processed/user_metrics.csv')
```

```python
# Convert date columns to datetime if they exist
date_columns = ['transaction_date', 'first_transaction', 'last_transaction']
for df in [processed_tx, user_metrics]:
    for col in date_columns:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col])
```

```python
# Display basic information about the data
print("Transaction data information:")
print(f"Number of transactions: {len(processed_tx)}")
print(f"Columns: {processed_tx.columns.tolist()}")
print("\nUser metrics information:")
print(f"Number of users: {len(user_metrics)}")
print(f"Columns: {user_metrics.columns.tolist()}")
```

```python
# Display a few rows of each dataset
print("\nSample transaction data:")
display(processed_tx.head())
print("\nSample user metrics:")
display(user_metrics.head())
```

```python
# Analyze transaction labels distribution
if 'label' in processed_tx.columns:
    label_counts = processed_tx['label'].value_counts()
    print("\nTransaction label distribution:")
    print(label_counts)
    
    # Visualize label distribution
    plt.figure(figsize=(10, 6))
    plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Transaction Type Distribution')
    plt.axis('equal')
    plt.show()
```

```python
# Analyze transaction volume over time
if 'time_step' in processed_tx.columns:
    tx_by_time = processed_tx.groupby('time_step').size()
    
    plt.figure(figsize=(12, 6))
    plt.plot(tx_by_time.index, tx_by_time.values, marker='o')
    plt.title('Transaction Volume Over Time')
    plt.xlabel('Time Step')
    plt.ylabel('Number of Transactions')
    plt.grid(True)
    plt.show()
```

```python
# Analyze transaction amounts
if 'total_input' in processed_tx.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(processed_tx['total_input'].clip(upper=processed_tx['total_input'].quantile(0.99)), bins=50)
    plt.title('Transaction Amount Distribution (99th Percentile)')
    plt.xlabel('Amount')
    plt.ylabel('Frequency')
    plt.show()
```
